**Building LogisticRegression model**

```python
class Logistic_Regression():

  #Initilizing the Parameters (Learning Rate and Number of Iterations)
  def __init__(self, learning_rate, no_of_iterations):
    self.learning_rate = learning_rate
    self.no_of_iterations = no_of_iterations

  def fit(self, X, Y):

    #Number of Training Examples(m) and Features (n) examples
    self.m, self.n = X.shape

    #Initialize the weights and bias
    self.w = np.zeros(self.n)
    self.b = 0
    self.X = X
    self.Y = Y

    #Implementing Gradient Descent for Optimizations
    for i in range(self.no_of_iterations):
      self.update_weights()

  def update_weights(self):

    #Y_hat formula for (Sigmoid Function)
    Y_hat = 1 / (1 + np.exp( -(self.X.dot(self.w) + self.b) )) #---> where z = WX + B

    #gradient descent deriviations with respect to W adn B
    dw = (1 / self.m) * np.dot(self.X.T, (Y_hat - self.Y))
    db = (1 / self.m) * np.sum(Y_hat - self.Y)

    #Updating weights and bias using gradient Descents algoritham
    self.w = self.w - self.learning_rate * dw
    self.b = self.b - self.learning_rate * db

  def predict(self, X):
    Y_pred = 1 / (1 + np.exp( -(X.dot(self.w) + self.b) ))
    Y_pred = np.where(Y_pred > 0.5, 1, 0)
    return Y_pred



```

#Implementing the LogisticRegression Model

```python
#importing the Dependencies
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#Loading the datasets and analysis
data = pd.read_csv('diabetes.csv')
```

#Seperating tha Features and Labels (Targets)

```python
X = data.drop(columns = 'Outcome', axis = 1)
Y = data['Outcome']
print(X)
print(Y)
```

#Spliting the data into Training and Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

#Standerdize the Features Data

```python
scaler = StandardScaler()
```

```python
x_train_stand = scaler.fit_transform(X_train)
x_test_stand = scaler.transform(X_test)
```

```python
print(x_train_stand)
```

```python
x = x_train_stand
y = Y_train
```

#Training the model

```python
classifier = Logistic_Regression(learning_rate = 0.01, no_of_iterations = 1000)
```

```python
#Training the LogisticRegression Model
classifier.fit(x,y)
```

#Model Evaluations

```python
#Accuracy Score on Training data
x_train_prediction = classifier.predict(x)
training_data_accuracy = accuracy_score(y, x_train_prediction)
print("Accuracy Score of training data:", training_data_accuracy)
```

```python
#Accuracy Score on Testing data
x_test_prediction = classifier.predict(x_test_stand)
testing_data_accuracy = accuracy_score(Y_test, x_test_prediction)
print("Accuracy Score of testing data:", testing_data_accuracy)
```
