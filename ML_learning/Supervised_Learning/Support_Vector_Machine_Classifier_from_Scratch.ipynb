```python
#Importing the Dependencies
import numpy as np
```

#Support Vector Machine Classifier

```python
class SVM_Classifier():

  #Initiating the hyperparameter
  def __init__(self, learning_rate, no_of_iterations, lambda_parameter):

    self.learning_rate = learning_rate
    self.no_of_iterations = no_of_iterations
    self.lambda_parameter = lambda_parameter

  #Fitting the Dataset to SVM Classifier
  def fit(self, X, Y):

    #Number of Training Examples(m) and Features (n) examples
    self.m, self.n = X.shape

    #Initialize the weights and bias
    self.w = np.zeros(self.n)
    self.b = 0
    self.X = X
    self.Y = Y

    #Implementing Gradient Descent for Optimizations
    for i in range(self.no_of_iterations):
      self.update_weights()


  #Function for Updating Weights and bias values
  def update_weights(self):

    #Label Encoding
    y_label = np.where(self.Y <=0, -1, 1)


    #Gradient (dw, db)
    for index, x_i in enumerate(self.X):

      condition = y_label[index] * (np.dot(x_i, self.w) - self.b) >=1

      if (condition == True):
        dw = 2 * self.lambda_parameter * self.w
        db = 0

      else:
        dw = 2 * self.lambda_parameter * self.w - np.dot(x_i, y_label[index])
        db = y_label[index]


      self.w = self.w - self.learning_rate * dw
      self.b = self.b - self.learning_rate * db


  #Predict the Label for given inputs
  def predict(self, X):

    output = np.dot(X, self.w) - self.b
    predicted_labels = np.sign(output)
    y_hat = np.where(predicted_labels <= -1, 0, 1)
    return y_hat



```

```python
#importing Dependencies
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

#Importing the data and preprocessing

```python
#Data collectiona and preprocessing
diabetes_data = pd.read_csv('diabetes.csv')
```

```python
#print the First 5 rows of Dataframe
diabetes_data.head()
```

```python
#Getting the Statistical Measurament
diabetes_data.describe()
```

```python
diabetes_data['Outcome'].value_counts()

# 0 ------> People with no Diabetes
# 1 ------> People with Diabetes
```

# Separting the Fetures and Target (labels)

```python
x = diabetes_data.drop(columns='Outcome', axis = 1)
y = diabetes_data['Outcome']
```

```python
print(x)
print(y)
```

#Split the data in Training and Testing Data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

#Standerdize the Data

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
X = x_train_scaled
Y = y_train
```

```python
print(X)
print(Y)
```

# Training the Model

```python
#Support Vector Machine Classifier
model = SVM_Classifier(learning_rate=0.001, no_of_iterations=1000, lambda_parameter=0.01)
```

```python
#Training the SVM Classifier model with Training Data
model.fit(X, Y)
```

#Model Evaluations

Accuracy Score

```python
#Accuracy Score on Training Data
X_training_prediction = model.predict(X)
training_data_accuracy = accuracy_score(Y, X_training_prediction)
print("Accuracy Score on Training Data:", training_data_accuracy)
```

```python
#Accuracy Score on Testing Data
X_testing_prediction = model.predict(x_test_scaled)
testing_data_accuracy = accuracy_score(y_test, X_testing_prediction)
print("Accuracy Score on Testing Data:", testing_data_accuracy)
```
